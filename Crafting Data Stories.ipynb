## Consumer Behavior and Shopping Habits Dataset

The data visualization showcases the purchase amount by category in the Consumer Behavior and Shopping Habits Dataset to visualize consumer preferences and purchasing behaviors. This visualization provides an overview of the categories that generate the highest purchase amounts, enabling businesses to optimize their product offerings and craft targeted marketing strategies. 

(https://www.kaggle.com/datasets/zeesolver/consumer-behavior-and-shopping-habits-dataset/?select=shopping_behavior_updated.csv)

```python
# Load package
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
shop_trend = pd.read_csv('shopping_trends.csv')
```

```python
print(shop_trend.head())
print(shop_trend.info())
```

```python
print(shop_trend.describe())
```

```python
# Visualization 2: Purchase Amount by Gender
plt.figure(figsize=(8, 6))
shop_trend.boxplot(column='Purchase Amount (USD)', by='Gender', grid=False)
plt.title('Purchase Amount by Gender')
plt.ylabel('Purchase Amount (USD)')
plt.xlabel('Gender')
plt.show()
```

Through a comparison of the boxplots for males and females, we can discern potential variations in the distribution of purchase amounts between the two genders. Specifically, the median purchase amount for males appears to be slightly higher than that of females. This discrepancy in median values implies a potential divergence in spending habits between males and females in this preliminary stage of analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded the 'shop_trend' DataFrame

# Calculate item counts by category
item_category_counts = shop_trend['Category'].value_counts()

# Calculate the total count of items purchased by each gender
total_counts_by_gender = shop_trend['Gender'].value_counts()

# Pivot the data to get the counts of items purchased by each gender and category
gender_category_counts = shop_trend.pivot_table(index='Category', columns='Gender', aggfunc='size', fill_value=0)

# Calculate the percentage of items purchased by category for each gender
percentage_by_category = (gender_category_counts / total_counts_by_gender) * 100

# Create the bar chart
plt.figure(figsize=(12, 6))
ax = percentage_by_category.plot(kind='bar', stacked=True, color=['lightcoral', 'lightskyblue'], width=0.7)
plt.title('Percentage of Items Purchased by Category and Gender', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Gender', loc='upper right', labels=['Female', 'Male'])

# Annotate the bars with percentage values
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.1f}%', (x + width / 2, y + height + 2), ha='center', fontsize=12, color='black')

plt.ylim(0, 100)  # Set the y-axis limit to 0-100 for percentages
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)  # Add a horizontal grid for better readability

plt.show()
```

Based on the above chart, it is evident that among males, accessories account for 32.0% of their purchases, followed by clothing at 44.5%, footwear at 15.1%, and outwear at 8.4%. On the other hand, for females, accessories make up 31.4% of their purchases, followed by clothing at 44.6%, footwear at 15.9%, and outwear at 8.1%.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded the 'shop_trend' DataFrame

# Calculate the total purchase amount by category
purchase_amount_by_category = shop_trend.groupby('Category')['Purchase Amount (USD)'].sum()

# Sort the categories in descending order
purchase_amount_by_category = purchase_amount_by_category.sort_values(ascending=False)

# Create the bar chart
plt.figure(figsize=(12, 6))
purchase_amount_by_category.plot(kind='bar', stacked=True, color='lightblue')
plt.title('Purchase Amount by Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Purchase Amount (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)  

# Show the plot
plt.show()
```

In this bar chart, it is evident that clothing has the highest purchase amount, followed by accessories, footwear, and outwear.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded the 'shop_trend' DataFrame

# Calculate the count of customers by gender
gender_counts = shop_trend['Gender'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightpink'])
plt.title('Gender Distribution', fontsize=16)
plt.ylabel('')

# Show the plot
plt.show()
```

The pie plot reveals that male shoppers account for 68.0% of the total, while female shoppers make up 32.0%.

```python
# Visualization 6: Subscription Status
subscription_counts = shop_trend['Subscription Status'].value_counts()
plt.figure(figsize=(8, 6))
subscription_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Subscription Status Distribution')
plt.show()
```

The pie chart illustrates the distribution of subscription status, with 73% of non-subscribers being male and 27% of subscribers being female.

```python
# Visualization 5: Purchase Amount Over Time (Season)
seasonal_purchase = shop_trend.groupby('Season')['Purchase Amount (USD)'].mean()
plt.figure(figsize=(8, 6))
seasonal_purchase.plot(kind='line', marker='o', color='blue')
plt.title('Average Purchase Amount by Season')
plt.xlabel('Season')
plt.ylabel('Average Purchase Amount (USD)')
plt.grid(True)
plt.show()
```

Based on the average purchase amount by season, it is evident that fall is the busiest shopping period. Conversely, summer is a slower month for sales. Interestingly, winter sales surpass the combined sales of both spring and summer.

```python
# Visualization 7: Preferred Payment Method
payment_method_counts = shop_trend['Preferred Payment Method'].value_counts()
plt.figure(figsize=(8, 6))
payment_method_counts.plot(kind='bar', color='lightcoral')
plt.title('Preferred Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Based on the bar chart depicting the preferred method of payment, it is evident that all payment methods are almost equally popular. However, PayPal and credit cards have a slight advantage over bank transfers, which are the least preferred option.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded the 'shop_trend' DataFrame

# Create the boxplot
plt.figure(figsize=(8, 6))
shop_trend.boxplot(column='Review Rating', by='Season', grid=False)
plt.title('Review Ratings by Season', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Review Rating', fontsize=12)

# Show the plot
plt.show()
```

In this box plot, which is grouped by season and reviews, it is apparent that the reviews are generally spread out or fairly consistent across all four seasons. The spring season shows a slightly higher median review compared to the other seasons. On the other hand, the summer season exhibits a greater variety or variance in the spread of reviews.
